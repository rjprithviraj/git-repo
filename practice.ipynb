```python
## forward and backward string out put
s = input("enter a string")
i = 0
n = len(s)

print ("Forward direction")
while i<n:
  print(s[i],end='')
  i = i+1

print()

print ("Backward direction")
i = -1
while i>=-n:
  print(s[i],end='')
  i = i-1
```
